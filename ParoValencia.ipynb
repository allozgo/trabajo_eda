```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importo el archivo con la población de cada municipio de Valencia

poblacion_valencia = pd.read_excel('Data/poblacion/poblacion_valencia.xls').drop(["Mun"], axis= 1)
poblacion_valencia
```

```python
# Reordeno las columnas

column_order = ['Codigo_postal', 'Municipio'] + [col for col in poblacion_valencia.columns if col not in ['Codigo_postal', 'Municipio']]
poblacion_valencia = poblacion_valencia[column_order]
poblacion_valencia
```

```python
# Saco las columnas que no son número para poder hacer su sumatorio, que me dará las cifras de población de la provincia de Valencia por año
# (como suma de las poblaciones de sus municipios)

poblacion_provincia = poblacion_valencia.iloc[:,10:]
poblacion_provincia.sum()
```

```python
# Hago la gráfica en seaborn de dicha población

plt.figure(figsize=(12,7))
sns.lineplot(data=poblacion_provincia.sum(),
            color = "purple",
            linewidth = 3)
plt.title("Población Total Provincia Valencia")
plt.xlabel("Año")
plt.ylabel("Millones de personas")
plt.show()
```

```python
# Saco la población en paro de la provincia de Valencia por municipio.  Como es un archivo por año con la información de todas las provincias, hago una iteración por todos los archivos
# y cojo la información que me interesa: de la provicia de Valencia de ese año.

provincias = ['PARO VALENCIA']

paro = pd.DataFrame()
aux2 = pd.DataFrame()

for i in range(2005, 2024, 1):
    if i == 2005:
        paro_valencia = pd.read_excel('Data/Data_paro/ESTADISTICA_MUNICIPIOS'+str(i)+'.xls', sheet_name='PARO VALENCIA').iloc[7:-1,1:3]
    else:
        if (i > 2012 or i < 2008): # Como estoy cogiendo sólo lo que necesito, hay unos años que las excel incluyen la firma del ministerio, que no me interesa
            aux2 = pd.read_excel('Data/Data_paro/ESTADISTICA_MUNICIPIOS'+str(i)+'.xls', sheet_name='PARO VALENCIA').iloc[7:-1,2:3]
        else:
            aux2 = pd.read_excel('Data/Data_paro/ESTADISTICA_MUNICIPIOS'+str(i)+'.xls', sheet_name='PARO VALENCIA').iloc[7:-3,2:3]
    paro_valencia = pd.concat([paro_valencia, aux2], axis = 1)
paro_valencia
```

```python
# Cambio el nombre de las columnas para saber de qué año se trata en cada caso

columnas = [ "Municipio", "2005", "2006", "2007", '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', "2023"]
paro_valencia.columns = columnas 
paro_valencia
```

```python
# Quito la columna municipio que no se puede convertir a float, paso las demás a ese formato y hago el sumatorio por columna, que me va a dar la población en paro
#  de la provincia por cada año

paro_total_valencia = paro_valencia.iloc[:,1:].astype(float)
paro_total_valencia.sum()
```

```python
# Hago la gráfica de dicha población en paro

plt.figure(figsize=(12,7))
sns.lineplot(data=paro_total_valencia.sum(),
            color = "cyan",
            linewidth = 3)
plt.title("Paro Total Provincia Valencia")
plt.xlabel("Año")
plt.ylabel("Personas")
plt.show()
```

```python
# Hago lo mismo que lo anterior, pero con una máscara para filtrar sólo los municipios con menos de 1000 habitantes

poblacion_valencia_1000 = poblacion_valencia[poblacion_valencia["2022"]<1000].iloc[:,11:]
poblacion_valencia_1000.sum()
```

```python
# Hago la gráfica del sumatorio de todas las columnas, la población en la provincia de los municipios de menos de 1000 habitantes, y por año.

plt.figure(figsize=(12,7))
sns.lineplot(data=poblacion_valencia_1000.sum(),
            color = "lightblue",
            linewidth = 3)
plt.title("Población en municipios < 1000 habitantes")
plt.xlabel("Año")
plt.ylabel("Personas")
plt.show()
```

```python
# Lo mismo, con el DataFrame de la población en paro

paro_val = paro_valencia.reset_index()[poblacion_valencia["2022"]<1000].iloc[:,2:]
paro_val = paro_val.astype(float)
```

```python
# Sumatorio por columnas para tener la población en paro en municipios de menos de 1000 habitantes

paro1000 = paro_val.sum()
paro1000
```

```python
# Y la correspondiente gráfica con dichos datos.

plt.figure(figsize=(12,7))
sns.lineplot(data=paro1000,
            color = "blue",
            linewidth = 3)
plt.title("Paro en municipios < 1000 habitantes")
plt.xlabel("Año")
plt.ylabel("Personas")
plt.show()
```

```python
# Igual que antes, con la máscara para los municipios con población entre 1000 y 5000 habitantes

poblacion_valencia_5000 = poblacion_valencia[(poblacion_valencia["2022"]>1000) & (poblacion_valencia["2022"]<5000)].iloc[:,11:]
poblacion_valencia_5000.sum()
```

```python
# La gráfica del sumatorio de datos de dicha población, cada columna un año.

plt.figure(figsize=(12,7))
sns.lineplot(data=poblacion_valencia_5000.sum(),
            color = "green",
            linewidth = 3)
plt.title("Población en municipios entre 1000 y 5000 habitantes")
plt.xlabel("Año")
plt.ylabel("Personas")
plt.show()
```

```python
# La población en paro en municipios de entre 1000 y 5000 habitantes

paro_val = paro_valencia.reset_index()[(poblacion_valencia["2022"]>1000) & (poblacion_valencia["2022"]<5000)].iloc[:,2:-1]
paro_val = paro_val.astype(float)
paro5000 = paro_val.sum()
paro5000
```

```python
# Su gráfica

plt.figure(figsize=(12,7))
sns.lineplot(data=paro5000,
            color = "lightgreen",
            linewidth = 3)
plt.title("Paro en municipios entre 1000 y 5000 habitantes")
plt.xlabel("Año")
plt.ylabel("Personas")
plt.show()
```

```python
# Esta vez la población de los minicipios de más de 5000 habitantes.

poblacion_valencia_mas5000 = poblacion_valencia[poblacion_valencia["2022"]>5000].iloc[:,11:]
poblacion_valencia_mas5000.sum()
```

```python
# Con su gráfica

plt.figure(figsize=(12,7))
sns.lineplot(data=poblacion_valencia_mas5000.sum(),
            color = "red",
            linewidth = 3)
plt.title("Población en municipios más de 5000 habitantes")
plt.xlabel("Año")
plt.ylabel("Personas")
plt.show()
```

```python
# Y los datos de población en paro en municipios de más de 5000 habitantes

paro_val = paro_valencia.reset_index()[poblacion_valencia["2022"]>5000].iloc[:,2:-1]
paro_val = paro_val.astype(float)
paro_mas_5000 = paro_val.sum()
paro_mas_5000
```

```python
# Con su gráfica

plt.figure(figsize=(12,7))
sns.lineplot(data=paro_mas_5000,
            color = "pink",
            linewidth = 3)
plt.title("Paro en municipios más de 5000 habitantes")
plt.xlabel("Año")
plt.ylabel("Personas")
plt.show()
```

```python
# Para poder hacer una comparativa mejor, hago el ratio dividiendo los datos de población en paro entre los datos totales de población.
# Para los municipios de menos de 1000 habitantes:

tasa_paro_1000 = paro1000 / poblacion_valencia_1000.sum()
tasa_paro_1000
```

```python
# Ratio para municipios de entre 1000 y 5000 habitantes

tasa_paro_5000 = paro5000 / poblacion_valencia_5000.sum()
tasa_paro_5000
```

```python
# Ratio para municipios con población mayor de 5000 habitantes.

tasa_paro_mas5000 = paro_mas_5000 / poblacion_valencia_mas5000.sum()
tasa_paro_mas5000
```

```python
# Y su correspondiente gráfico

plt.figure(figsize=(12,7))
sns.lineplot(data=tasa_paro_1000,
            color = "blue",
            linewidth = 3)
sns.lineplot(data=tasa_paro_5000,
            color = "green",
            linewidth = 3)
sns.lineplot(data=tasa_paro_mas5000,
            color = "red",
            linewidth = 3)
plt.title("Comparativa Ratio paro según tamaño municipio")
plt.xlabel("Año")
plt.ylabel("Ratio")
plt.show()
```
